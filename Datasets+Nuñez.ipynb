# Dataset #1: *household_income*

## Descripción del dataset

Corresponde a un dataset con varios factores demográficos y socioeconomicos que influencian el ingreso familiar anual. Las columnas que componen el dataset se describen a continuación.

* Age: Edad del jefe de hogar (18 to 70 years).

* Education Level: Nivel de educación más alto (High School, Bachelor's, Master's, Doctorate).

* Occupation: Tipo de ocupación (Healthcare, Education, Technology, Finance, Others).

* Number of Dependents: Numero de personas dependientes del jefe de hogar (0 to 5).

* Location: Tipo de zona residencial (Urban, Suburban, Rural).

* Work Experience: Años de experiencia laboral (0 to 50 years).

* Marital Status: Estado civil del jefe de hogar (Single, Married, Divorced).

* Employment Status: Situacion laboral del jefe de hogar (Full-time, Part-time, Self-employed).

* Household Size: Numero total de miembros del hogar (1 to 7).

* Homeownership Status: Situación de posesión de la vivienda (Own, Rent).

* Type of Housing: Tipo de vivienda (Apartment, Single-family home, Townhouse).

* Gender: Género del jefe de hogar (Male, Female).

* Primary Mode of Transportation: Modo de transporte del jefe de hogar (Car, Public transit, Biking, Walking).

* Annual Household Income: Ingreso anual del hogar en USD

# Dataset #2: *spotify_songs_2024*

## Descripción del dataset

Corresponde a un dataset con varios datos sobre el rendimiento de canciones de spotify en diversas plataformas, incluyendo spotify, youtube, tiktok, etc. Las columnas que componen el dataset se describen a continuación.

* Track Name: Nombre de la canción.
* Album Name: Nombre del album.
* Artist: Nombre del artista.
* Release Date: Fecha de lanzamiento.
* ISRC: Codigo de grabación estandar internacional (International Standard Recording Code).
* All Time Rank: Ranking de la canción basada en su popularidad global.
* Track Score: Puntuación asignada a la cancion basada en varios factores.
* Spotify Streams: Numero total de streans en spotify.
* Spotify Playlist Count: Numero de playlist que incluyen a la cancion.
* Spotify Playlist Reach: Alcance de la cancion a traves de las playlists de spotify.
* Spotify Popularity: Puntaje de popularidad de la cancion en spotify.
* YouTube Views: Numero total de vistas del video oficial de la cancion en youtube.
* YouTube Likes: Numero total de likes del video oficial de la cancion en youtube.
* TikTok Posts: Numero total de posteos de TikTok que contienen la cancion.
* TikTok Likes: Numero total de likes de los posteos de TikTok que contienen la cancion.
* TikTok Views: Numero total de vistas de los posteos de TikTok que contienen la cancion.
* YouTube Playlist Reach: Alcance de la cancion a traves de las playlists de YouTube.
* Apple Music Playlist Count: Numero de playlists de Apple Music que contienen la cancion.
* AirPlay Spins: Numero de veces que la cancion ha sido reproducida en estaciones de radio.
* SiriusXM Spins: Numero de veces que la cancion ha sido reproducida en SiriusXM.
* Deezer Playlist Count: Numero de playlists de Deezer que contienen la cancion.
* Deezer Playlist Reach: Alcance de la cancion a traves de las playlists de Deezer.
* Amazon Playlist Count: Numero de playlists de Amazon Music que contienen la cancion.
* Pandora Streams: Numero total de transmisiones en Pandora.
* Pandora Track Stations: Numero total de estaciones de Pandora que han transmitido la cancion.
* Soundcloud Streams: Numero total de transmisiones en Soundcloud.
* Shazam Counts: Numero total de veces que la cancion se ha Shazameado.
* TIDAL Popularity: Puntaje de popularidad de la cancion en TIDAL.
* Explicit Track: Indica si la cancion contiene lenguaje explicito.
